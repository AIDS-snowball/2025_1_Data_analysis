## **< 중복된 행 찾기 >**
- **duplicated() 메서드** : True - 중복된 행 중 처음 행을 제외한 나머지 행, False - 중복되지 않은 행 \
- 열을 기준으로 중복 찾음
- ex) 중복된 행 있는지 확인

```python
sum(ns_book.duplicated())
```

**subset 매개변수**
- 일부 열을 기준으로 중복된 행 찾음\
- ex) 도서명, 저자, ISBN 기준으로 중복된 행 찾음 -> 2만개 넘는 중복 존재

```python
sum(ns_book.duplicated(subset=['도서명', '저자','ISBN']))
```

**keep 매개변수**
- duplicated() 메서드에 keep 매개변수를 False로 지정 -> 중복된 모든 행 True로 표시 \
 -> ns_book3 변수에 저장

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
#'파친코'는 두 권짜리 도서, '보건교사 안은영'은 두 권 등록됨
```

**< 그룹별로 모으기 >**
- **groupby() 메서드** : **by 매개변수** 이용 -> 행을 합칠 때 기준이 되는 열 지정
- 중복된 행 삭제 : _drop_duplicates() 메서드_ 사용
- ex) 그룹으로 묶을 기준 열과 '대출권수'열 선택하여 그룹화 \

```python
count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
```

-> count_df 데이터프레임에 **groupby() 메서드** 적용 \
-> 정수 타입인 '대출건수'열은 **sum() 메서드** 사용 \
-> groupby() 메서드는 **NaN**이 포함되어있으면 행을 삭제함 \
-> 이를 막기 위해 dropna 매개변수를 False로 지정 -> NaN 행도 포함


```python
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False)
loan_count = group_df.sum()
```

```python
# groupby() 메서드와 sum() 메서드 연이어 호출 -> 어떤 작업을 수행하는지 확인 가능
loan_count = count_df.groupby(by=['도서명', '저자','ISBN', '권'],
                              dropna=False).sum()
loan_count.head()
```

## **< 원본 데이터 업데이트하기 >**
1. duplicated() 메서드로 중복된 행을 True로 표시한 불리언 배열을 만듭니다
2. 1번에서 구한 불리언 배열을 반전시켜서 중복되지 않은 고유한 행을 True로 표시합니다
3. 2번에서 구한 불리언 배열을 사용해 원본 배열에서 고유한 행만 선택합니다 \
- 중복된 행을 True로 표시한 불리안 배열을 반전시킬 때 **~연산자** 사용
- 원본 배열에서 고유한 배열을 선택하여 **copy() 메서드**로 ns_book3 데이터프레임 만든다

```python
dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
unique_rows = ~dup_rows
ns_book3 = ns_book[unique_rows].copy()
```

**copy() 메서드는 왜 사용하나요?**
_ copy() 메서드는 복사본을 만든다. 업데이트 할 때마다 데이터를 저장할 복사본은 만든다

```python
#ns_book3 중복된 행 없는지 확인
sum(ns_book3.duplicated(subset=['도서명', '저자','ISBN', '권']))
```

## **원본 데이터프레임 인덱스 설정하기**
- **set_index() 메서드** : 지정한 열을 인덱스로 설정
- **inplace 매개변수** : True로 지정하여 새로운 데이터프레임이 아닌 원본 데이터프레임 인덱스 수정
- ex) 인덱스가 변경된 ns_book3 데이터프레임의 처음 다섯 개 행을 출력\
=> loan_count 데이터프레임과 동일하게 '도서명', '저자', 'ISBN', '권' 열이 인덱스가 됨

```python
ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
ns_book3.head()
```

##**업데이트하기: update() 메서드**
- 업데이트 후 인덱스 열 해제
- **rest_index() 메서드**로 데이터프레임 인덱스 재설정 가능

```python
ns_book3.update(loan_count)
ns_book3.head()
```

## **일괄 처리 함수 만들기**
- 열 삭제 : Ioc 메서드, 슬라이싱, 불리언 배열, drop() 메서드
- 행 삭제 : [ ] 연산자, 슬라이싱, 불리언 배열
- 중복된 행 합치기 : groupby() 메서드, sum()메서드 \
=> 원본 데이터프레임 업데이트 위해 update() 메서드 사용\
=> 이 때 인덱스 기준으로 업데이트 하기 위해 set_index() 메서드, reset_index() 메서드
- **일괄 처리 함수** : 작성했던 코드를 새로운 데이터로 적용하기 쉽도록 일괄 처리하는 함수

```python
def data_cleaning(filename):
  ns_df = pd.read_csv(filename, low_memory=False)
  ns_book = ns_df.dropna(axis=1, how='all')
  count_df = ns_book[['도서명', '저자', 'ISBN', '권', '대출건수']]
  loan_count = count_df.groupby(by=['도서명', '저자', 'ISBN', '권'], dropna=False).sum()
  dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN', '권'])
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()
  ns_book3.set_index(['도서명', '저자', 'ISBN', '권'], inplace=True)
  ns_book3.update(loan_count)
  ns_book4 = ns_book3.reset_index()
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4
```

- **equals() 메서드** : 다른 데이터프레임과 동일한지 비교

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4)
# 두 데이터프레임이 동일한지 비교 -> 완전히 동일
```

- **데이터 정제** \
: 잘못된 부분을 수정, 제거하여 필요한 데이터를 준비하는 과정. 데이터가 올바르게 정제 X -> 분석된 결과 왜곡 가능성 O

- **데이터 랭글링 or 데이터 먼징** \
: 데이터 정제하는 과정, 데이터 분석 및 머신러닝에 적합한 형태로 데이터를 변환하는 과정

- **원소별 비교** -> 비교한 결과를 True, False로 이루어진 **불리언 배열**로 반환

- **넘파이** : 다차원 배열, 한 종류의 데이터만 담을 수 있음

- DataFrame.drop() : 행 or 열 삭제
- DataFrame.dropna() : 누락된 값이 포함된 행 or 열 삭제
- DataFrame.duplicated() : 중복된 행을 찾아 불리언 값으로 표시한 배열을 반환
- DataFrame.groupby() : 데이터프레임의 행을 그룹으로 모음
- DataFrame. sum() : 행 or 열 기준으로 합계 계산
- DataFrame.set_index : 지정한 열을 인덱스로 설정
- DataFrame.reset_index() : 데이터프레임의 인덱스 재설정
- DataFrame.update() : 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트, 다른 데이터프레임에 있는 NaN은 업데이트에서 제외
- DataFrame.equals() : 다른 데이터프레임과 동일한 원소를 가졌는지 비교 -> 동일하면 True, 아니면 False반환
