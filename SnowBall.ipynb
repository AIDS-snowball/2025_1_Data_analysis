```python
import pandas as pd

df = pd.read_csv('WineQT.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

columns = df.columns
n = int(len(columns) ** 0.5) + 1 #4*4를 만들기 위한 계산
fig, axs = plt.subplots(n, n, figsize=(10, 10))

for i in range(n):
  for j in range(n):
    idx = i * n + j
    if idx < len(columns):
      axs[i][j].hist(df[columns[idx]])
      axs[i][j].set_title(columns[idx])
    else:
      axs[i][j].axis('off')
```

```python
mean_by_quality = df.groupby(by='quality').mean()
mean_by_quality
```

```python
import matplotlib.pyplot as plt

columns = mean_by_quality.columns
n = int(len(columns) ** 0.5) + 1 #4*4를 만들기 위한 계산
fig, axs = plt.subplots(n, n, figsize=(10, 10))

for i in range(n):
  for j in range(n):
    idx = i * n + j
    if idx < len(columns):
      axs[i][j].bar(mean_by_quality.index, mean_by_quality.iloc[:, idx])
      axs[i][j].set_title(columns[idx])
    else:
      axs[i][j].axis('off')
```
