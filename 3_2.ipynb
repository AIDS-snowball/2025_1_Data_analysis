## **4. 숫자 찾기: \d**
- 정규 표현식에서 숫자를 나타내는 기호 : **\d**
- 네 자리 연도 : **\d\d\d\d**
- 그룹으로 묶기 -> 괄호 : 뒤에 두자리만 묶기 **\d\d(\d\d)**
- 패턴 안에 있는 그룹 : \1,\2 ... (등장순서 순으로)
- **regex 매개변수** = True : 정규표현식 사용


```python
#'발행년도' 열의 값을 정규 표현식으로 두 자리 연도로 바꿈
# 패턴 안에 있는 첫 번째 그룹으로 연도를 바꿈
ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]
```

**r 문자** : 정규 표현식을 다른 문자열과 구분하기 위함\
**중괄호** : 정규 표현식 반복될 때 개수 지정 \
ex) \d\d = \d{2} -> 연속된 숫자 두 개\
ex) \d{2}(\d{2}) -> 앞에서 사용한 숫자 네 개

```python
# \d\d(\d\d) = \d{2}(\d{2})
ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

- **gt() 메서드** : 전달된 값보다 큰 값을 찾음

```python
#연도가 4000년이 넘는 행의 전체 개수 확인을 위해 sum()함수 함께 사용
ns_book5['발행년도'].gt(4000).sum()
```

**< gt()메서드 대신 () 기호를 쓸 수 있음 >** \
ns_book5['발행년도'].gt(4000).sum() \
= (ns_book5['발행년도'] > 4000).sum() \

데이터 프레임 원소 비교할 수 있는 메서드\
gt() : > : 지정된 값보다 큰 값 검사\
ge() : >= : 지정된 값보다 크거나 같은 값 검사\
It() : < : 지정된 값보다 작은 값을 검사\
le() : <= : 지정된 값보다 작거나 같은 값을 검사\
eq() : == : 지정된 값과 같은 값을 검사\
ne() : != : 지정된 값과 같지 않은 값을 검사

```python
#누락되거나 알 수 없는 행을 뷰티플수프를 사용해 채우기
import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
  # URL에 ISBN을 넣어 HTML 가져옵니다.
  url = 'http://www.yes24.com/Product/Search?domain=B00K&query={}'
  r = requests.get(url.format(isbn))
  # HTML 파싱
  soup = BeautifulSoup(r.text, 'html.parser')
  # 클래스 이름이 'gdjame'인 <a> 태그의 텍스트를 가져옵니다.
  title = soup.find('a', attrs={'class': 'gd_name'}).get_text()
  return title
```

- **find_all() 메서드** : 저자를 담은 \<a> 태그 모두 추출
- **for문** : \<a> 태그에 속한 모든 텍스트를 파이썬 리스트에 저장
-**join() 메서드** : 하나의 문자열로 합침
- **re모듈, findall() 함수** : 원하는 정규식에 매칭되는 모든 문자열을 찾아 리스트로 반환

```python
import re

def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub= row['출판사']
  year= row['발행년도']
  #Yes24 도서 검색 페이지 URL
  url='http://www.yes24.com/Product/Search?domain=B00K&query={}'
  #URL에 ISBN을 넣어 HTML 가져옵니다
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser') #HTML파싱
  try:
    if pd.isna(title):
      #클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옵니다
      title = soup.find('a', attrs={'class':'gd_name'}).get_txt()
  except AttributeError:
    pass

  try:
    if pd.isna(author):
      #클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옵니다
      authors = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
      author_list = [auth.get_text() for auth in authors]
      author = ','.join(author_list)
  except AttributeError:
    pass

  try:
    if pd.isna(pub):
    #클래스 이름이 'info_pub'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옵니다
      pub = soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()
  except ArithmeticError:
    pass

  try:
    if year==-1:
      #클래스 이름이 'info_date'인 <span> 태그 아래 텍스트를 가져옵니다.
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      #정규 표현식으로 찾은 값 중에 첫 번째 값만 사용한다.
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  return title, author, pub, year
```

- 뷰티플수프로 안될 경우 오류 발생\
\-> 다음 요소 추출하도록 **try ~ except문**으로 예외 처리


- **result_type 매개변수**로 반환된 값을 각기 다른 열로 만듦

##**8. 데이터를 이애하고 올바르게 정제하기**

```python
def data_fixing(ns_book4):

#잘못된 값을 수정하거나 NaN을 채우는 함수
# :param ns_book4: data一cleaning。함수에서 전처리된 데이터프레임

# 도서권수와 대출건수를 int32로 바꿉니다.
  ns_book4 = ns_book4.astype({'도서권수':'int32', '대출건수': 'int32'})
 # NaN인 세트 ISBN을 빈 문자열로 바꿉니다.
  set_isbn_na_rows = ns_book4['세트 ISBN'].isna()
  ns_book4.loc[set_isbn_na_rows, '세트 ISBN'] = ''
 # 발행년도 열에서 연도 네 자리를 추출하여 대체합니다. 나머지 발행년도는 -1 로 바꿉니다.
  ns_book5 = ns_book4.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex=True)
  unkown_year = ns_book5['발행년도'].str.contains('\D', na=True)
  ns_book5.loc[unkown_year, '발행년도'] = '-1'
 # 발행년도를 int32로 바꿉니다.
  ns_book5 = ns_book5.astype({'발행년도': 'int32'})
 # 4000년 이상인 경우 2333년을 뺍니다.
  dangun_yy_rows = ns_book5['발행년도'] .gt(4000)
  ns_book5.loc[dangun_yy_rows, '발행년도'] = ns_book5.loc[dangun_yy_rows,\
                                                      '발행년도'] - 2333
# 여전히 4000년 이상인 경우 -1 로 바꿉니다.
  dangun_year = ns_book5 ['발행년도'].gt (4000)
  ns_book5.loc[dangun_year, '발행년도'] = -1
 # 0-1900^ 사이의 발행년도는 -1 로 바꿉니다.
  old_books = ns_book5['발행년도'].gt(0) & ns_book5['발행년도'].lt(1900)
  ns_book5.loc[old_books, '발행년도'] = -1
 # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1 인 행을 찾습니다.
  na_rows = ns_book5['도서명'].isna() | ns_book5['저자'].isna() \
  | ns_book5['출판사'].isna() | ns_book5['발행년도'].eq(-1)
 # Yes24 도서 상세 페이지에서 누락된 정보를 채웁니다.
  updated_sample = ns_book5[na_rows].apply(get_book_info,
                                           axis=1, result_type = 'expand')
  # 아래 줄의 들여쓰기를 수정했습니다.
  updated_sample.columns = ['도서명','저자','출판사','발행년도']
  ns_book5.update(updated_sample)
 # 도서명, 저자, 출판사가 NaN이거나 발행년도가 -1 인 행을 삭제합니다.
  ns_book6 = ns_book5.dropna(subset=['도서명','저자','출판사'])
  ns_book6 = ns_book6[ns_book6['발행년도'] != -1]

  return ns_book6
```

  isna()메서드 사용 -\> NaN 여부 확인,\
  notna()메서드 사용 -\> NaN이 아닌 값 체크
