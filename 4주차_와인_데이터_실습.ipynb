```python
import matplotlib.pyplot as plt

plt.scatter(aaa['density'], aaa['alcohol'])
plt.xlabel('density')   #x좌표 density
plt.ylabel('alcohol')    #y좌표 alcohol
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(aaa['density'], aaa['pH'])
plt.xlabel('density')   #x좌표 density
plt.ylabel('pH')    #y좌표 pH
plt.show()
```

```python
plt.hist(aaa['quality'])


plt.xlabel('Quality')       # x축: 와인 등급
plt.ylabel('Frequency')     # y축: 해당 등급 개수

plt.title('Quality Distribution')

plt.show()
```

```python
plt.hist(aaa['quality'], bins=20)
plt.yscale('log')
plt.show()
```

```python
plt.boxplot(aaa[['quality']])
plt.yscale('log')

plt.show()
```

```python
plt.boxplot(aaa[['quality']], vert=False)
plt.xscale('log')
plt.show()
```

```python
#산점도
import matplotlib.pyplot as plt

plt.scatter(df['quality'], df['alcohol'], alpha=0.6)
plt.xlabel('qualtiy')   #x좌표 quality
plt.ylabel('alcohol')        #y좌표 alcohol
plt.title('qualtiy vs alcohol')

plt.show()
```

```python
import matplotlib.pyplot as plt


plt.plot(df['quality'], df['pH'], alpha=0.6)


plt.xlabel('quality')
plt.ylabel('pH')

plt.show()
```
